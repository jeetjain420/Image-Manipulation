```python
# using histogram equalized image
img_norm = img.copy()
meanval = np.mean(img_norm.ravel())
stdval = np.std(img_norm.ravel())
img_norm = (img_norm - meanval) / stdval
plt.imshow(np.uint8(img_norm))
```

```python
plt.hist(np.uint8(img_norm.ravel(), bins=50))
```
